```python
# crie um programa para mostrar as 4 operações da tabuada e print o resultado de cada um
for i in range(1,11):
    for k in range (1,11):
        print(f'{i} x {k} = {i*k}')
        print(f'{i} / {k} = {i/k}')
        print(f'{i} + {k} = {i+k}')
        print(f'{i} - {k} = {i-k}')
```

```python
lista_desorganizada = [3,7,4,5,2,1]

def organizador(lista):
    lista_organizada = []
    for i in range(len(lista)):
        for k in range(len(lista)):
            if lista[i] < lista[k]: 
                lista[k] = lista[i]
                if lista[i] in lista_organizada:
                    continue
                lista_organizada.append(lista[k])
                
                
    return lista_organizada


print(organizador(lista_desorganizada))
```
