```python
""" desenvolva um algoritmo para calcular a area de um circulo e retornar as seguintes informações: 
a) area > 100 - circulo grande, b) area > 50, c) area < 50 - circulo pequeno, utilizando função dentro de função.

"""

def area_circulo(raio):
    area = 3.14 * raio ** 2
    return area

def mostrar_resultado_raio(raio):
    area = area_circulo(raio)
    if area > 100:
        print("Circulo grande")
    elif area > 50:
        print("Circulo médio")
    else:
        print("Circulo pequeno")
        
raio = float(input("Digite o raio do circulo: "))
mostrar_resultado_raio(raio)
```

```python
""" desenvolva um algoritmo para calcular a soma e a media de uma lista, utilize duas funções"""
def funcsoma(lista):
    soma = sum(lista)
    return soma
def funcmedia(lista):
    media = sum(lista)/len(lista)
    return media

def mostrar_resultado_soma_media_lista(lista):
    soma = funcsoma(lista)
    media = funcmedia(lista)
    
    return soma, media

lista = [1, 2, 3, 4, 5]
soma, media = mostrar_resultado_soma_media_lista(lista)
print(soma)
print(media)
```

```python
""" Crie uma função para retornar a soma e a media da lista1 = [10,20,30,40,50], uma função para retornar a soma e a media da 
lista2 = [5,10,20,40,80] e a função final, a qual retornará a média da soma e a média final das duas anteriores.
"""

lista1 = [10,20,30,40,50]

lista2 = [5,10,20,40,80]

def sumavelist(lista: list)->float: 
    soma = sum(lista)
    media = sum(lista)/len(lista)
    
    return soma, media

def mediafinal(lista1: list, lista2: list)->float:
    soma1, media1 = sumavelist(lista1)
    soma2, media2 = sumavelist(lista2)
    media_da_soma= soma1 + soma2/2
    
    mediafinal = (media1 + media2)/2
    
    return media_da_soma, mediafinal

aaaaaa, bbbbbb = mediafinal(lista1, lista2)
print(aaaaaa)
print(bbbbbb)

```

```python
"""Crie uma calculadora simples com uma função para cada operação, A função deve ter 2 parametros, os dois parametros devem ser enviados
por meio de um append para o resultado"""


def calculadora(num1: float,num2: float)->float:
    operacao = str(input('selecione entre: +,-,*,/'))
    resultado = []
    
    def soma(num1: float, num2: float) -> float:
        resultado = num1 + num2
        return resultado
    def subtracao(num1: float, num2: float) -> float:
        resultado = num1 - num2
        return resultado
    def divisao(num1: float, num2: float)-> float:
        resultado = num1 / num2
        return resultado
    def multiplicacao(num1: float, num2: float)-> float:
        resultado = num1 * num2
        return resultado
    
    match operacao:
        case '+':
            resultado.append(soma(num1, num2))
        case '-':
            resultado.append(subtracao(num1, num2))
        case '*':
            resultado.append(multiplicacao(num1, num2))
        case '/':
            resultado.append(divisao(num1, num2))
        case _:
            print('Operação inválida')
    return resultado

calculadora(5,5)
    
```
