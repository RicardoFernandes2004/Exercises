```python
# Crie um programa para calcular a média ponderada de 4 valores de vendas com entrada de dados pelo usuário. 
# Em seguida, exiba a porcentagem de cada valor de venda em relação à média ponderada

valor1 = float(input("Insira o valor 1: "))
peso1 = float(input("Insira o peso 1: "))
valor2 = float(input("Insira o valor 2: "))
peso2 = float(input("Insira o peso 2: "))
valor3 = float(input("Insira o valor 3: "))
peso3 = float(input("Insira o peso 3: "))
valor4 = float(input("Insira o valor 4: "))
peso4 = float(input("Insira o peso 4: "))


soma_valores = valor1 + valor2 + valor3 + valor4
soma_pesos = peso1 + peso2 + peso3 + peso4
media_ponderada = (valor1 * peso1 + valor2 * peso2 + valor3 * peso3 + valor4 * peso4) / soma_pesos

print(f"A média ponderada é: {media_ponderada}")

percentual1 = (valor1 * peso1 / soma_valores) * 100
percentual2 = (valor2 * peso2 / soma_valores) * 100
percentual3 = (valor3 * peso3 / soma_valores) * 100
percentual4 = (valor4 * peso4 / soma_valores) * 100

print(f"O valor {valor1} representa {percentual1}% da média ponderada.")
print(f"O valor {valor2} representa {percentual2}% da média ponderada.")
print(f"O valor {valor3} representa {percentual3}% da média ponderada.")
print(f"O valor {valor4} representa {percentual4}% da média ponderada.")

```
